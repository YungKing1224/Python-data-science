# **Spam detection: Naive Bayes**
The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research. It contains one set of SMS messages in English of 5,574 messages, tagged acording being ham (legitimate) or spam.

```python
import pandas as pd
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
df.head()
```

```python
df.columns
```

As we can see Unnamed 2 to 4 are not supposed to be part of the data set so we will drop them.

```python
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis='columns', inplace = True)
df.head()
```

I will rename the columns v1 to Category and v2 to Message for better understanding of the data.

```python
df.rename(columns={'v1':'Category', 'v2':'Message'}, inplace=True)
df.head()
```

Check for null values.

```python
df.isna().sum()
```

We are now splitting the data into train test datasets and setting the test size to 25%

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.Category, test_size = 0.25)
```

For this project we will use sklearn learn pipeline to convert the messages into count vectorizer. Then it will use Multinomial Naive Bayes to train the data set.

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
