# 1. Import data using Pandas
read_csv function is used to import the csv data and then stored in as a data frame in df. Which is printed at the end.

```python
import pandas as pd
```

```python
df = pd.read_csv(r'BostonHousing.csv')
df
```

# 2. Prepare data
## Data cleaning
Data with blank values are saved as NaN. we need to remove those blank values. dropna function is being used. axis = 0 means we want to remove all rows with NaN.

```python
df.dropna(axis = 0, inplace = True)
df
```

## Alternative data cleaning
To do data cleaning use another df and do not add inplace=True. Using inplace=True modifies the DataFrame df directly.

```python
df2 = df.dropna(axis=0)
df2
```

# 3. Perform basic statistical analysis
## Print max age column

```python
df['age']

max_age = max(df['age'])
print(max_age)
```

## Print min age column

```python
min_age = min(df['age'])
print(min_age)
```

## Import numpy
To get mean we have to import numpy as pandas does not support the mean function.

```python

import numpy as np
```

## Print mean age column

```python
mean_age = np.mean(df['age'])
print(mean_age)
```

## Alternative way to get statistical results
1. Count - Counts the number of observations
2. Mean - The average value
3. Std - Standard deviation (explained in the statistics chapter)
4. Min - The lowest value
5. 25%, 50% and 75% are percentiles (explained in the statistics chapter)
6. Max - The highest value
