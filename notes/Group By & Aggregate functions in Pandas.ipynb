```python
import pandas as pd
```

```python
df = pd.read_csv(r'/Users/Jason/Downloads/Flavors.csv')
df
```

```python
##group by base flavor
group_by_frames = df.groupby('Base Flavor')
```

```python
##calculate mean
##add numeric only to avoid error
group_by_frames.mean(numeric_only=True)
```

```python
##count different variables with the row we aggregated which are chocolate and vanilla.
group_by_frames.count()
```

```python
##Calculate min
group_by_frames.min()
```

```python
##Calculate max
group_by_frames.max()
```

```python
##total values
group_by_frames.sum()
```

```python
##there is an aggregation function to get for a specific column
group_by_frames.agg({'Flavor Rating':['min', 'mean', 'max', 'count']})
```

```python
##for multiple columns
group_by_frames.agg({'Flavor rating':['min', 'mean', 'max', 'count', 'sum'], 'Texture Rating': ['min', 'mean', 'max', 'count', 'sum']})
```
