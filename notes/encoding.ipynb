```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
le = LabelEncoder()
```

```python
# Use the 'raw' link to get the CSV data
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv"

df = pd.read_csv(url)
df
```

```python
df2 = df
df2['Car Model'] = le.fit_transform(df2['Car Model'])
df2
```

```python
x = df2.drop(['Sell Price($)'], axis='columns')
x
```

```python
y = df2['Sell Price($)']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder()

# Use ColumnTransformer to specify which columns to apply the encoder to
column_transformer = ColumnTransformer(
    [('one_hot_encoder', ohe, [0])],    # [0] specifies the column index
    remainder='passthrough'             # Keep the other columns unchanged
)

# Fit and transform the data
transformed_data = column_transformer.fit_transform(x)

print(transformed_data)
```

```python
transformed_data = transformed_data[:,1:]
transformed_data
```

```python
model = LinearRegression()
```

```python
model.fit(transformed_data, y)
```

```python
model.score(transformed_data, y)
```
