```python
import pandas as pd
```

```python
df = pd.read_csv(r'BostonHousing.csv')
df
```

```python
##print max age column
df['age']

max_age = max(df['age'])
print(max_age)
```

```python
min_age = min(df['age'])
print(min_age)
```

```python
##mean is found using numpy
import numpy as np
```

```python
##use np to get mean
mean_age = np.mean(df['age'])
print(mean_age)
```

```python
##get top 5 rows using head
df.head()
```

```python
# data cleaning
# data with blank values are saved as NaN. we need to remove those blank values
#dropna function is being used. axis = 0 means we want to remove all rows with NaN.
df.dropna(axis = 0, inplace = True)
df
```

```python
##to do data cleaning use another df and do not add inplace=True
#Using inplace=True modifies the DataFrame df directly
df2 = df.dropna(axis=0)
df2
```

```python
##use info function to get data types in dat set
df2.info()
```

```python
##in case there is a different data type such as object which cannot be calculated or get analysis performed
##use a astype function. This changes age column into float data type
df2['age'] = df2['age'].astype(float)
```

```python
# Count - Counts the number of observations
# Mean - The average value
# Std - Standard deviation (explained in the statistics chapter)
# Min - The lowest value
# 25%, 50% and 75% are percentiles (explained in the statistics chapter)
# Max - The highest value
df2.describe()
```
