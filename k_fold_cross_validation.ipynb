```python
import pandas as pd
from sklearn.datasets import load_iris
iris =load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, )
df.head()
```

```python
df['target']= iris.target
df.head()
```

```python
x = df.drop(['target'], axis='columns')
x.head()
```

```python
y = df.target
y.head()
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lr, x, y)
```

```python
mean_score_lr = cross_val_score(lr, x, y).mean()
print(f'Logistic Regression mean score is: {mean_score_lr}')
```

```python
cross_val_score(rfc,x,y)
```

```python
mean_score_rfc = cross_val_score(rfc,x,y).mean()
print(f'Random Forest Classifier mean score is: {mean_score_rfc}')
```

```python
cross_val_score(dtc, x, y)
```

```python
mean_score_dtc =cross_val_score(dtc, x, y).mean()
print(f'Decision Tree Classifier mean score is: {mean_score_dtc}')
```

Use a function to print mean scores of all the models stored in the list models. This is a more efficient way of coding rather than printing one at a time.

```python
models = [lr, rfc, dtc]

def get_score(x, y):
    for model in models:
        score = cross_val_score(model, x, y).mean()
        print(f'Mean score of {model.__class__.__name__} is: {score}')

get_score(x, y)
```
